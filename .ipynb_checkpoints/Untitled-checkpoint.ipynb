```python
job_data = pd.read_csv('jobs.csv')
```

```python
job_data
```

```python
Wellington = job_data[job_data['location'].str.contains(pat = 'Wellington', na=False)]
Wellington
```

```python
Auckland = job_data[job_data['location'].str.contains(pat = 'Auckland', na=False)]
Auckland['Posted_date'] = pd.to_datetime(Auckland['Posted_date'], format='%d/%m/%y')
Auckland
```

```python

grouped = Auckland.groupby(Auckland['Posted_date'].dt.date).count()
grouped.rename(columns={'Posted_date': 'date'}, inplace=True)
grouped.reset_index(inplace=True)
grouped = grouped[['Posted_date', 'title']]
grouped.columns = ['date', 'job_count']
grouped.sort_values(by='date', ascending=True, inplace=True)
grouped
```

```python
plt.bar(grouped['date'], grouped['job_count'])
plt.xlabel('Date')
plt.ylabel('Job Count')
plt.xticks(rotation=45)
plt.show()
```
